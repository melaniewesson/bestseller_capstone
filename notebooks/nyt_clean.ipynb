```python
import pandas as pd
```

```python
nyt = pd.read_csv(r"C:\Users\melan\Documents\DA8\Projects\bestseller_capstone\data\merged_dataframe.csv")
```

```python
nyt.head()
```

```python
#drop contributor_note, price, age_group
nyt = nyt.drop(['contributor', 'contributor_note', 'price', 'age_group', 'asterisk', 'dagger'], axis=1)
nyt.head()
```

```python
nyt['title'] = nyt['title'].str.title()
nyt.head()
```

```python
len(nyt)
```

```python
nyt['primary_isbn13'].nunique()
```

```python
#Add columnn for max weeks on list for each isbn
nyt['max_weeks_on_list'] = nyt.groupby('primary_isbn13')['weeks_on_list'].transform('max')
nyt.head()
```

```python
nyt.nlargest(1, 'max_weeks_on_list')
```

```python
# Group the DataFrame by the book's identifier and count the occurrences
book_counts = nyt.groupby('title').apply(lambda x: (x['rank'] == 1).sum())

# Find the book with the highest count
most_common_book = book_counts.idxmax()
count = book_counts.loc[most_common_book]

print("The book that appeared most in the #1 spot is:", most_common_book)
print("Number of times in the #1 spot:", count)
```

```python
#read in goodreads csv

goodreads = pd.read_csv(r"C:\Users\melan\Documents\DA8\Projects\bestseller_capstone\data\clean_goodreads.csv")
```

```python
goodreads.head()
```

```python
len(goodreads)
```

```python
nyt['primary_isbn13'] = nyt['primary_isbn13'].astype(str)
goodreads['isbn'] = goodreads['isbn'].astype(str)
```

```python
bestsellers = nyt.merge(goodreads, left_on='primary_isbn13', right_on='isbn', how='left')
```

```python
bestsellers.head()
```

```python
bestsellers.columns
```

```python
bestsellers = bestsellers.drop(['title_y', 'author_y', 'Unnamed: 0', 'isbn'], axis=1)
bestsellers = bestsellers.rename(columns={'title_x':'title', 'author_x':'author'})
bestsellers.head()
```
