```python
#turn unique_to_scrape into a list to loop through
#making pairs from nyt title, and last name of nyt author
title_author_pairs = list(zip(unique_to_scrape['title_x'], unique_to_scrape['author_x'].str.split().str[-1]))
title_author_pairs
```

```python
#make everything lower case
title_author_pairs = [(title.lower(), author.lower()) for title, author in title_author_pairs]
title_author_pairs
```

```python
### current issue, serach is working, but isn't clicking any element at all

# Write loop

# Initialize empty lists
goodreads_found = []
goodreads_unavailable = []

# Set the timeout value
timeout = 10
max_retries = 3

# Iterate over the list
for title, author in title_author_pairs[:10]:
    found = False
    retry_count = 0
    while retry_count < max_retries and not found:
        try:
            # Look up title in the search bar
            search_book = driver.find_element(By.NAME, "q")
            search_book.clear()  # Clear previous search
            search_book.send_keys(title)
            search_book.submit()

            # Wait for results to load
            wait = WebDriverWait(driver, timeout)
            search_results_loaded = wait.until(EC.presence_of_element_located((By.TAG_NAME, "a")))

            # Find the link with the matching title and author
            link_elements = driver.find_elements(By.TAG_NAME, "a")

            # Iterate through link elements on the search page
            for link_element in link_elements:
                # Find the span element containing the book title
                title_element = link_element.find_element(By.TAG_NAME, "span")
                link_title = title_element.text.strip()

                # Check if the link's title contains any excluded keywords
                excluded_keywords = ['summary', 'hardback', 'paperback', 'trivia', 'Hardcover', 'quiz', 'antique', 'sidekick', 'infographic', 'illustrated', 'facts', 'lesson plan', 'for fans', 'collection', 'author', 'conversation', 'analysis', 'review', 'synopsis', 'study', 'discussion', 'journal', 'notes', 'workbook', 'signed', 'preview']
                if any(keyword in link_title.lower() for keyword in excluded_keywords):
                    continue  # Skip this link element if it contains excluded keywords

                # Find the span element containing the author name
                author_element = link_element.find_element(By.CSS_SELECTOR, "span[itemprop='author'] span[itemprop='name']")
                link_author = author_element.text.strip().lower()

                # Check if the link's title and author match the desired values
                if title.lower() in link_title.lower() and author.lower() in link_author:
                    # Click on the link element
                    link_element.click()
                    found = True
                    break

            if found:
                # Wait for the book page to load
                book_page_loaded = wait.until(EC.presence_of_element_located((By.XPATH, f"//span[contains(text(), '{author.split()[-1]}')]")))

                # Make soup
                soup = BeautifulSoup(driver.page_source, 'html.parser')
                gr_title = soup.find('h1', {'class': 'Text Text__title1'}).text.strip() if soup.find('h1', {'class': 'Text Text__title1'}) else None
                gr_author = soup.find('span', {'class': 'ContributorLink__name'}).text.strip() if soup.find('span', {'class': 'ContributorLink__name'}) else None
                avg_rating = soup.find('div', {'class': 'RatingStatistics__rating'}).text.strip() if soup.find('div', {'class': 'RatingStatistics__rating'}) else None
                count_rating = soup.find('span', {'data-testid': 'ratingsCount'}).text.strip().split()[0].replace(',', '') if soup.find('span', {'data-testid': 'ratingsCount'}) else None
                count_reviews = soup.find('span', {'data-testid': 'reviewsCount'}).text.strip().split()[0].replace(',', '') if soup.find('span', {'data-testid': 'reviewsCount'}) else None
                date_published = soup.find('p', {'data-testid': 'publicationInfo'}).text.strip() if soup.find('p', {'data-testid': 'publicationInfo'}) else None

                # Scrape the genre
                genre_elements = soup.select('div.BookPageMetadataSection__genres > ul > span > span')
                genres = [genre.text.strip() for genre in genre_elements]
                genre1 = ', '.join(genres) if genres else None

                # Append scraped data to goodreads_found list
                goodreads_found.append({
                    'title': gr_title,
                    'author': gr_author,
                    'avg_rating': avg_rating,
                    'count_rating': count_rating,
                    'count_reviews': count_reviews,
                    'date_published': date_published,
                    'top_genre': genre1
                })

            break

        except Exception as e:
            retry_count += 1
            print(f"Exception occurred: {e}. Retrying...")

     
    else:
        #Not on goodreads
        goodreads_unavailable.append({
            'title': title,
            'author': author
        })

if retry_count == max_retries:
    print(f"Max retries exceeded for book: {title}, {author}")
```
