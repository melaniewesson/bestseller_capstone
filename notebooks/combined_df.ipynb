```python
from pathlib import Path
import pandas as pd
import numpy as np
```

```python
#set working directory
path = r"C:\Users\melan\Documents\DA8\Projects\BestsellersCapstone\capstone\data\final"
#get file names
files = Path(path).glob('*csv')
```

```python
dfs = list()
for f in files:
    data = pd.read_csv(f)
    data['file'] = f.stem
    dfs.append(data)


df = pd.concat(dfs, ignore_index=True)
```
